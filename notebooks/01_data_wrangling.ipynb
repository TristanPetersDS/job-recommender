```python
import os, sys

import pandas as pd
import numpy as np

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
```

```python
DATA_PATH = "../data/raw/"
PLOT_PATH = "../outputs/"
```

```python
#jobs_df = pd.read_csv(os.path.join(DATA_PATH, 'jobs.csv'))
post_df = pd.read_csv(os.path.join(DATA_PATH, 'postings.csv'))
resume_df = pd.read_csv(os.path.join(DATA_PATH, 'Resume.csv'))

job_skills = pd.read_csv(os.path.join(DATA_PATH, 'jobs/job_skills.csv'))
mapped_skills = pd.read_csv(os.path.join(DATA_PATH, 'mappings/skills.csv'))
```

```python
filtered_df = post_df.dropna(subset=['title', 'description', 'location', 'skills_desc', 'formatted_experience_level'])
filtered_df.info()
```

```python
filtered_df.head() 
```

```python
merged_skills = job_skills.merge(mapped_skills, on='skill_abr', how='left')
job_id = post_df['job_id']

filtered_skills = merged_skills[merged_skills['job_id'].isin(job_id)]

skill_text = filtered_skills.groupby('job_id')['skill_name'].apply(lambda x: ' '.join(x)).reset_index()
skill_text.columns = ['job_id', 'skills_text']

filtered_df = filtered_df.merge(skill_text, on='job_id', how='left')
filtered_df['skills_text'] = filtered_df['skills_text'].fillna('')
```
